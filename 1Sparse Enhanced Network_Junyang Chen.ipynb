# 2.GAN（生成对抗网络）

GAN的主要应用目标：

生成式任务（生成、重建、超分辨率、风格迁移、补全、上采样等）

GAN的核心思想：生成器G和判别器D的一代代博弈

生成器：生成网络，通过输入生成图像

判别器：二分类网络，将生成器生成图像作为负样本，真实图像作为正样本

learn 判别器D：

给定G，通过G生成图像产生负样本，并结合真实图像作为正样本来训练D

learn 生成器G：

给定D，以使得D对G生成图像的评分尽可能接近正样本作为目标来训练G

G和D的训练过程交替进行，这个对抗的过程使得G生成的图像越来越逼真，D“打假”的能力也越来越强。



- 例子：可以看作一个假币机不断通过随机取样数据生成的假币和判别机不断对生成的假币进行检验的生成对抗过程

## 1. 传统推荐系统 vs 顺序推荐系统

### 传统推荐系统：
- 基于用户和物品的特征或者行为数据。
- 主要依赖于用户对物品的评分、点击次数等静态信息。
- 例子：用户A喜欢科幻电影，系统会推荐更多的科幻电影给用户A。

### 顺序推荐系统：
- 关注用户行为的时间顺序。
- 考虑用户的行为序列，捕捉用户兴趣的变化和行为模式。
- 例子：用户B先看了一部科幻电影，然后连续看了几部动作电影，系统会推荐更多的动作电影而不是科幻电影，因为用户的兴趣可能已经转移。

## 2. 为什么要用顺序推荐模型？
- **动态兴趣捕捉**：用户的兴趣随时间变化，顺序推荐模型可以更好地捕捉这种变化。
- **上下文关联**：用户的行为有时具有上下文关联性，比如看完一部电影后通常会看相关的续集或者同类型的电影。
- **更准确的预测**：考虑行为顺序可以提高推荐的准确性，使推荐更贴近用户当前的需求和兴趣。

## 3. 顺序推荐模型的工作原理
顺序推荐模型主要通过分析用户行为序列来做出推荐。常见的方法包括：

- **基于序列的协同过滤**：利用用户行为序列之间的相似性来进行推荐。
- **序列模式挖掘**：发现用户行为序列中的模式，如频繁项集等。
- **时间序列模型**：使用如隐马尔可夫模型（HMM）或长期短期记忆网络（LSTM）等模型来捕捉时间序列特征。
- **深度学习方法**：采用诸如RNN（循环神经网络）、Transformer等深度学习模型来处理和预测用户行为序列。

## 4. 举个例子
假设有一个视频平台，用户在一段时间内的观看记录如下：

- 第一天：看了《哈利·波特1》
- 第二天：看了《哈利·波特2》
- 第三天：看了《哈利·波特3》

## 5. 问题

- **实时性要求**：需要对用户的行为变化做出快速反应。

# InfoNCE损失

一旦定义好了正负样本对，就需要一个对比学习的损失函数来指导模型来进行学习。这个损失函数需要满足这些要求，

即当 query q 和唯一的正样本相似，并且和其他所有负样本key都不相似的时候，这个loss的值应该比较低。反之，如果和不相似，或者和其他负样本的key相似了，那么loss就应该大，从而惩罚模型，促使模型进行参数更新。


文中公式


#### 正样本对 (Positive Pair):
$h_{a_i}^u$ 和 $h_{a_j}^u$ 是通过不同数据增强操作 $a_i$ 和 $a_j$ 生成的同一用户 $u$ 的序列表示。这些表示被认为是相似的（正样本对）。
#### 负样本对 (Negative Pair):
$h_{\bar{a_j}}^u$ 是其他用户 $\bar{u}$ 的增强序列表示，视为负样本对。它们与正样本对应该有较大的差异。
#### 相似度函数:
$\text{sim}(\cdot)$ 通常表示为点积（dot product），用于衡量两个向量之间的相似度。相似度值越高，向量越相似。
$\tau$ 是温度参数（temperature parameter），用于调节点积的数值，以控制相似度分数的尺度。
#### 分子部分:
$\exp(\text{sim}(h_{a_i}^u, h_{a_j}^u)/\tau)$ 计算正样本对之间的相似度，期望该值尽可能大。
#### 分母部分:
$\sum_{\bar{u} \in \text{neg}} \exp(\text{sim}(h_{a_i}^u, h_{\bar{a_j}}^u)/\tau)$ 计算所有负样本对的相似度之和，期望该值尽可能小。
通过对所有负样本对进行求和，确保模型能够区分正样本对和负样本对。
#### 损失函数:
通过最大化正样本对的相似度（使分子更大）和最小化负样本对的相似度（使分母更大），InfoNCE 损失函数实现了对比学习的目标。
最终的损失函数值是所有样本的对数相似度和的负值，模型通过最小化这个损失来优化其参数。

如果缩放因子τ 设的越大，logits分布变得越平滑，那么对比损失会对所有的负样本一视同仁，导致模型学习没有轻重。

# 真正理解点积的意义(之前只会公式)



1. **投影**：点积可以用于计算一个向量在另一个向量方向上的投影。
2. **物理意义**：在物理学中，点积表示两个向量之间的平行分量的积，例如，计算一个力在一个方向上的做功。


**！** **相似性度量**：在机器学习和信息检索中，点积常用于衡量两个向量的相似性。例如，在推荐系统中，用户和物品的特征向量之间的点积可以用来表示用户对物品的偏好程度。


```python
# 导入必要的库
import numpy as np
```

```python
# 定义两个向量
a = np.array([1, 2, 3])
b = np.array([1, 2,3])
c = np.array([-1, -2,-3])
```

```python
# 计算点积
dot_product_plus = np.dot(a, b)
dot_product_neg = np.dot(a, c)
```

```python
# 打印结果
print(f"向量 a: {a}")
print(f"向量 b: {b}")
print(f"向量 a 和 b 的点积: {dot_product_plus}")
print(f"向量 a 和 c 的点积: {dot_product_neg}")
```

```python
# 在推荐系统中的应用
# 假设我们有用户和物品的特征向量
user_vector = np.array([0.1, 0.3, 0.5])
item_vector = np.array([0.2, 0.4, 0.6])
```

```python
# 计算用户和物品特征向量的点积
preference_score = np.dot(user_vector, item_vector)
```

```python
# 打印结果
print(f"用户特征向量: {user_vector}")
print(f"物品特征向量: {item_vector}")
print(f"用户对物品的偏好分数: {preference_score}")
```

点积可以用来衡量向量之间的相似性。向量的点积越大，表明它们的方向越相近，从而相似性越高。

## 指标的意义

1. **Hit@K (Hit Rate at K)**:
    - **定义**：在前 K 个推荐结果中，用户感兴趣的项目（命中）的比例。
    - **意义**：衡量推荐系统推荐的项目中，用户实际感兴趣的项目的比例。Hit@K 越高，说明推荐系统的推荐效果越好。

2. **NDCG@K (Normalized Discounted Cumulative Gain at K)**:
    - **定义**：NDCG 衡量的是推荐结果的排序质量，考虑了推荐项目的相关性和排序位置。NDCG@K 是在前 K 个推荐结果中的 NDCG 值。
    - **意义**：衡量推荐系统推荐结果的相关性和排序质量。NDCG@K 越高，说明推荐系统不仅推荐了用户感兴趣的项目，而且推荐的项目排在了较前的位置。

# HIT 和 NDCG 指标计算公式

## HIT（Hit Rate）
HIT（命中率）是一个评价推荐系统准确性的指标，用来衡量推荐结果中是否包含用户真正感兴趣的物品。HIT@K表示在推荐的前K个结果中，至少有一个结果是用户感兴趣的。

- **公式**：HIT@K = $\frac{\text{Number of hits}}{\text{Number of users}}$

- **含义**：如果系统向用户推荐了K个物品，其中至少有一个物品是用户实际点击或购买的，那么就称为一次命中。HIT@K即为在所有用户中有多少次命中。

## NDCG（Normalized Discounted Cumulative Gain）
NDCG（归一化折损累积增益）是一个综合考虑推荐结果相关性和排序的评价指标。它不仅考虑推荐结果是否相关，还考虑推荐结果的排序位置。NDCG@K表示在推荐的前K个结果中的NDCG值。

- **公式**：
  $
  DCG@K = \sum_{i=1}^{K} \frac{rel_i}{\log_2(i+1)}
  $
  其中，$rel_i$表示第i个结果的相关性得分。
  
  NDCG@K 是对DCG@K进行归一化，归一化的标准是理想情况下的DCG（即最相关的结果在最前面）：
  
  $
  NDCG@K = \frac{DCG@K}{IDCG@K}
  $

  其中，IDCG@K是理想情况下的DCG值。

- **解释**：NDCG考虑了结果的相关性和排序位置。相关性高且排在前面的结果贡献更高的分数。因此，NDCG能够更全面地评估推荐结果的质量。

## 栗子
假设有一个用户喜欢的物品集合为{A, B, C}，推荐系统推荐了物品{D, A, B}。

- 对于HIT@3，如果推荐结果中包含用户喜欢的物品（如A或B），则认为是一次命中。此例中，推荐结果包含A和B，所以HIT@3为1（表示命中）。
- 对于NDCG@3，首先计算DCG：
  $
  DCG@3 = \frac{1}{\log_2(2)} + \frac{1}{\log_2(3)} + \frac{0}{\log_2(4)} = 1 + 0.6309 + 0 = 1.6309
  $

  然后计算理想情况下的IDCG（A和B排在最前面）：
  $
  IDCG@3 = \frac{1}{\log_2(2)} + \frac{1}{\log_2(3)} = 1 + 0.6309 = 1.6309
  $
  
  因此，NDCG@3 = \(\frac{1.6309}{1.6309} = 1\)。

## SASRecModel 的基本结构

SASRecModel 的基本结构可以分为以下几个部分：

1. **嵌入层（Embedding Layer）**：
   - 用户和物品的嵌入层，用于将离散的用户和物品 ID 映射到连续的向量表示。
   - 位置嵌入层，用于表示序列中每个位置的信息。

2. **编码器（Encoder）**：
   - 基于 Transformer 的编码器，使用自注意力机制来捕捉序列中元素之间的依赖关系。

3. **预测层（Prediction Layer）**：
   - 将编码后的表示通过全连接层转换为评分，用于预测用户对物品的偏好。

**该文件定义了如下函数**

9. **Intermediate**：前馈神经网络层。
10. **Layer**：Transformer编码器层，包括自注意力和前馈神经网络。
11. **Encoder**：Transformer编码器，由多个Layer组成。
