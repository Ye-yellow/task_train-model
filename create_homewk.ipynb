# 整合过程

```python
import pandas as pd



# 读取文件
social_network_path = 'E:\\video\KuaiRec 2.0\data\social_network.csv'
item_categories_path = 'E:\\video\KuaiRec 2.0\data\item_categories.csv'
big_matrix_path = 'E:\\video\KuaiRec 2.0\data\\big_matrix.csv'
item_daily_features_path = 'E:\\video\KuaiRec 2.0\data\item_daily_features.csv'
small_matrix_path = 'E:\\video\KuaiRec 2.0\data\small_matrix.csv'



item_categories_df = pd.read_csv(item_categories_path)

item_daily_features_df = pd.read_csv(item_daily_features_path)
small_matrix_df = pd.read_csv(small_matrix_path)
social_network_df = pd.read_csv(social_network_path)

```

```python
# 读取原始的Beauty.txt数据文件以获取用户-视频交互格式
beauty_path = 'E:\work11\SparseEnNet-main\SparseEnNet\data\Beauty.txt'
with open(beauty_path, 'r') as f:
    beauty_data = f.readlines()

# 解析原始数据
user_sequences = {}
for line in beauty_data:
    parts = line.strip().split()
    user_id = int(parts[0])
    video_ids = list(map(int, parts[1:]))
    user_sequences[user_id] = video_ids

```

```python
# 生成新的用户-视频交互数据 
for index, row in small_matrix_df.iterrows():
    user_id = int(row['user_id'])
    video_id = int(row['video_id'])
    if user_id not in user_sequences:
        user_sequences[user_id] = []
    user_sequences[user_id].append(video_id)

```

```python
# 将数据保存为homewk.txt格式
output_path = './homewk.txt'
with open(output_path, 'w') as f:
    for user_id, video_ids in user_sequences.items():
        line = f"{user_id} " + " ".join(map(str, video_ids)) + "\n"
        f.write(line)

print(f"整合后的数据已保存至 {output_path}")
```
